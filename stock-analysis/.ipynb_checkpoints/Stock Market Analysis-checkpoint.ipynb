## Time Series Data

A time series is a sequence of numerical data points taken at successive equally spaced points in time.
In investing, a time series tracks the movement of the stock price, over a specified period of time.

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

%matplotlib inline
```

```python
plt.plot(amzn.index, amzn['Adj Close'])

plt.show()
```

```python
plt.plot(amzn.index, amzn['Adj Close'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.grid(True)
plt.xticks(rotation=90)

plt.show()
```

### Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(10,10), sharex=True)
fig.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
fig.gca().xaxis.set_major_locator(mdates.YearLocator())

axes[0,0].plot(amzn.index, amzn['Adj Close'], color='r')
axes[0,0].grid(True)
axes[0,0].tick_params(labelrotation=90)
axes[0,0].set_title('AMZN');

axes[0,1].plot(fb.index, fb['Adj Close'], color='g')
axes[0,1].grid(True)
axes[0,1].tick_params(labelrotation=90)
axes[0,1].set_title('FB');

axes[1,0].plot(goog.index, goog['Adj Close'], color='b')
axes[1,0].grid(True)
axes[1,0].tick_params(labelrotation=90)
axes[1,0].set_title('GOOG');

axes[1,1].plot(aapl.index, aapl['Adj Close'], color='y')
axes[1,1].grid(True)
axes[1,1].tick_params(labelrotation=90)
axes[1,1].set_title('AAPL');
```

### Zooming-in

```python
# Sublots
fig, axes = plt.subplots(2, 2, figsize=(30,25), sharex=True, sharey=True)
fig.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
fig.gca().xaxis.set_major_locator(mdates.MonthLocator())

amzn_19_20 = amzn.loc[pd.Timestamp('2019-01-01'):pd.Timestamp('2020-10-01')]
axes[0,0].plot(amzn_19_20.index, amzn_19_20['Adj Close'], '.', color='r')
axes[0,0].grid(True)
axes[0,0].tick_params(labelrotation=90)
axes[0,0].set_title('AMZN');

fb_19_20 = fb.loc[pd.Timestamp('2019-01-01'):pd.Timestamp('2020-10-01')]
axes[0,1].plot(fb_19_20.index, fb_19_20['Adj Close'], '.', color='g')
axes[0,1].grid(True)
axes[0,1].tick_params(labelrotation=90)
axes[0,1].set_title('FB');

goog_19_20 = goog.loc[pd.Timestamp('2019-01-01'):pd.Timestamp('2020-10-01')]
axes[1,0].plot(goog_19_20.index, goog_19_20['Adj Close'], '.', color='b')
axes[1,0].grid(True)
axes[1,0].tick_params(labelrotation=90)
axes[1,0].set_title('GOOG');

aapl_19_20 = aapl.loc[pd.Timestamp('2019-01-01'):pd.Timestamp('2020-10-01')]
axes[1,1].plot(aapl_19_20.index, aapl_19_20['Adj Close'], '.', color='y')
axes[1,1].grid(True)
axes[1,1].tick_params(labelrotation=90)
axes[1,1].set_title('AALP');
```

```python
monthly_amzn_19_20 = amzn_19_20.resample('4M').mean()
plt.scatter(monthly_amzn_19_20.index, monthly_amzn_19_20['Adj Close'])
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=90)
plt.show()
```

```python
# Subplots
fix, axes = plt.subplots(2, 2, figsize=(10,10), sharex=True, sharey=True)

monthly_amzn_19_20 = amzn_19_20.resample('4M').mean()
axes[0,0].scatter(monthly_amzn_19_20.index, monthly_amzn_19_20['Adj Close'], color='r')
axes[0,0].grid(True)
axes[0,0].tick_params(labelrotation=90)
axes[0,0].set_title('AMZN');

monthly_fb_19_20 = fb_19_20.resample('4M').mean()
axes[0,1].scatter(monthly_fb_19_20.index, monthly_fb_19_20['Adj Close'], color='g')
axes[0,1].grid(True)
axes[0,1].tick_params(labelrotation=90)
axes[0,1].set_title('FB');

monthly_goog_19_20 = fb_19_20.resample('4M').mean()
axes[1,0].scatter(monthly_goog_19_20.index, monthly_goog_19_20['Adj Close'], color='b')
axes[1,0].grid(True)
axes[1,0].tick_params(labelrotation=90)
axes[1,0].set_title('GOOG');

monthly_aapl_19_20 = fb_19_20.resample('4M').mean()
axes[1,1].scatter(monthly_aapl_19_20.index, monthly_aapl_19_20['Adj Close'], color='y')
axes[1,1].grid(True)
axes[1,1].tick_params(labelrotation=90)
axes[1,1].set_title('AAPL');
```

### Analysing Difference between Levels (Resampling Weekly)

```python
plt.scatter(amzn_diff.loc['2020-01-01':'2020-10-01'].index, amzn_diff.loc['2020-01-01':'2020-10-01']['Difference'])
# plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=90)
plt.show()
```

```python
# Subplot

fb['Difference'] = fb['Open'] - fb['Close']
fb_diff = fb.resample('W').mean()

goog['Difference'] = goog['Open'] - goog['Close']
goog_diff = goog.resample('W').mean()

aapl['Difference'] = aapl['Open'] - aapl['Close']
aapl_diff = aapl.resample('W').mean()


f, ax = plt.subplots(2, 2, figsize=(10,10), sharex=True, sharey=True)

ax[0,0].scatter(amzn_diff.loc['2020-01-01':'2020-10-01'].index, amzn_diff.loc['2020-01-01':'2020-10-01']['Difference'], color='r')
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('AMZN');

ax[0,1].scatter(fb_diff.loc['2020-01-01':'2020-10-01'].index, fb_diff.loc['2020-01-01':'2020-10-01']['Difference'], color='g')
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('FB');

ax[1,0].scatter(goog_diff.loc['2020-01-01':'2020-10-01'].index, goog_diff.loc['2020-01-01':'2020-10-01']['Difference'], color='b')
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('GOOG');

ax[1,1].scatter(aapl_diff.loc['2020-01-01':'2020-10-01'].index, aapl_diff.loc['2020-01-01':'2020-10-01']['Difference'], color='y')
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('AAPL');
```

## Step 3: Moving Windows

### Daily Percentages

```python
# AMZN
daily_close_amzn = amzn[['Adj Close']]

# Daily returns
daily_pct_changes_amzn = daily_close_amzn.pct_change()

# Replace NA value with 0
daily_pct_changes_amzn.fillna(0, inplace=True)

daily_pct_changes_amzn
```

```python
daily_pct_changes_amzn.hist(bins=50)

plt.show()
```

```python
# FB
daily_close_fb = fb[['Adj Close']]

daily_pct_changes_fb = daily_close_fb.pct_change()

daily_pct_changes_fb.fillna(0, inplace=True)


daily_pct_changes_fb.hist(bins=50)

plt.show()
```

```python
# GOOG
daily_close_goog = goog[['Adj Close']]

daily_pct_changes_goog = daily_close_goog.pct_change()

daily_pct_changes_goog.fillna(0, inplace=True)


daily_pct_changes_goog.hist(bins=50)

plt.show()
```

```python
# AAPL
daily_close_aapl = aapl[['Adj Close']]

daily_pct_changes_aapl = daily_close_aapl.pct_change()

daily_pct_changes_aapl.fillna(0, inplace=True)


daily_pct_changes_aapl.hist(bins=50)

plt.show()
```

```python
import seaborn as sns

sns.set()
```

```python
# Set up the matplotlib figure
f, axes = plt.subplots(2, 2, figsize=(12, 7))

# Plot a simple histogram with binsize determined automatically
sns.distplot(daily_pct_changes_amzn['Adj Close'], color='r', ax=axes[0,0], axlabel='AMZN');

# Plot a kernel density estimate and rug plot
sns.distplot(daily_pct_changes_fb['Adj Close'], color='g', ax=axes[0,1], axlabel='FB');

# Plot a kernel density estimate and rug plot
sns.distplot(daily_pct_changes_goog['Adj Close'], color='b', ax=axes[1,0], axlabel='GOOG');

# Plot a kernel density estimate and rug plot
sns.distplot(daily_pct_changes_aapl['Adj Close'], color='y', ax=axes[1,1], axlabel='AAPL');
```

## Step 4: Volatality

```python
import numpy as np
```

```python
min_periods = 75

# Calculate the volatility
vol = daily_pct_changes_amzn.rolling(min_periods).std() * np.sqrt(min_periods)

vol.fillna(0, inplace=True)

vol
```

```python
# Plot the volatility
vol.plot(figsize=(10, 8))

plt.show()
```

## Rolling Means (Trends and Seasonality)

```python
# AMZN
amzn_adj_close_px = amzn['Adj Close']

# Short moving window rolling mean
amzn['42'] = amzn_adj_close_px.rolling(window=42).mean()

# Lont moving window rolling mean
amzn['252'] = amzn_adj_close_px.rolling(window=252).mean()

# Plot the adjust closing price, the short and long windows of rolling means
amzn[['Adj Close', '42', '252']].plot(title='AMZN')

plt.show()
```

```python
# FB
fb_adj_close_px = fb['Adj Close']

# Short moving window rolling mean
fb['42'] = fb_adj_close_px.rolling(window=42).mean()

# Lont moving window rolling mean
fb['252'] = fb_adj_close_px.rolling(window=252).mean()

# Plot the adjust closing price, the short and long windows of rolling means
fb[['Adj Close', '42', '252']].plot(title='FB')

plt.show()
```

```python
# GOOG
goog_adj_close_px = goog['Adj Close']

# Short moving window rolling mean
goog['42'] = goog_adj_close_px.rolling(window=42).mean()

# Lont moving window rolling mean
goog['252'] = goog_adj_close_px.rolling(window=252).mean()

# Plot the adjust closing price, the short and long windows of rolling means
goog[['Adj Close', '42', '252']].plot(title='GOOG')

plt.show()
```

```python
# AAPL
aapl_adj_close_px = goog['Adj Close']

# Short moving window rolling mean
aapl['42'] = aapl_adj_close_px.rolling(window=42).mean()

# Lont moving window rolling mean
aapl['252'] = aapl_adj_close_px.rolling(window=252).mean()

# Plot the adjust closing price, the short and long windows of rolling means
aapl[['Adj Close', '42', '252']].plot(title='AAPL')

plt.show()
```
