```python
# Remove stopwords and remove short words (less than 2 characters)
def preprocess(text):
  result = []
  for token in gensim.utils.simple_preprocess(text):
    if token not in stop_words and len(token) >= 3:
      result.append(token)
  return result
```

# TASK #7: PREPARE THE DATA BY TOKENIZING AND PADDING

![alt text](https://drive.google.com/uc?id=13j8m-JOpK994CtukR1EShiY_hGGjkNx-)

```python
stock_df
```

```python
# Obtain the total words present in the dataset

list_of_words = []

for i in stock_df['Test Without Punc & Stopwords']:
  for j in i:
    list_of_words.append(j)
```

```python
list_of_words
```

```python
# Obtain the total number of unique words
total_words = len(list(set(list_of_words)))
total_words
```

```python
# split the data into test and train 
X = stock_df['Test Without Punc & Stopwords']
y = stock_df['Sentiment']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Create a tokenizer to tokenize the words and create sequences of tokenized words
tokenizer = Tokenizer(num_words=total_words)
tokenizer.fit_on_texts(X_train)

# Training Data
train_sequences = tokenizer.texts_to_sequences(X_train)

# Testing Data
test_sequences = tokenizer.texts_to_sequences(X_test)
```

```python
train_sequences
```

```python
test_sequences
```

```python
print('The encoding for document\n', X_train[1:2], 'is: ', train_sequences[1])
```

```python
# Add padding to training and testing
padded_train = pad_sequences(train_sequences, maxlen=29)
padded_test = pad_sequences(test_sequences, maxlen=29)
```

```python
for i, doc in enumerate(padded_train[:3]):
  print('The padded encoding for document:', i+1, 'is', doc)
```

```python
# Convert the data to categorical 2D representation
y_train_cat = to_categorical(y_train, 2)
y_test_cat = to_categorical(y_test, 2)
```
